# Matplotlib

### Matplotlib Basics

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y = x*2
```

```python
plt.plot(x,y) # plot np.array x and y
plt.title("X vs Y") # give the title
plt.xlabel("X Axis") # x label
plt.ylabel("Y Axis") # y label
plt.xlim(0, 6) # limit the axis
plt.ylim(0, 6)
plt.savefig("myfirstplot.png") # save the plot as png
plt.show() # show the plot
```

### Matplotlib - Understanding the Figure Object

Matplotlib makes use of a Figure Object.
We then add axes to this Figure Object and the plot on those axes.

```python
fig = plt.figure(figsize=(10,5)) # create the Figure object 10 inches x 5 inches
axes = fig.add_axes([0,0,1,1]) # add the axes (0,0) lower left corner, (1,1) width and hight of Axes 1 means 100% of the figure
axes.plot(x,y)
plt.title("X vs Y") # give the title
plt.xlabel("X Axis") # x label
plt.ylabel("Y Axis") # y label
plt.show()
```

```python
a = np.linspace(0,10,11)
b = a**4
```

### Matplotlib - Figure Parameters

```python
fig = plt.figure(figsize=(8,4), dpi=200) #dpi = Dot for Inch
axes1 = fig.add_axes([0,0,1,1]) 
axes1.plot(a,b)
fig.savefig("mysecondplot.png", bbox_inches="tight") # bbox_inches="tight" needed to show the axes in the png

plt.show()
```

### Matplotlib - Subplots Functionality

```python
fig, axes = plt.subplots(nrows=2, ncols=2) # create 4 plots 
axes[0][1].plot(a,b) # plot on the top, right set of axes
axes[1][0].plot(x,y) # plot on the bottom, left set of axes
plt.tight_layout() # to improve the gap between the subplots
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) # create 4 plots 
for ax in axes:
    ax.plot(a,b) # we can also loop throgh the subplots
plt.tight_layout() # to improve the gap between the subplots
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2) 
axes[0][1].plot(a,b)
axes[1][0].plot(x,y) 
fig.subplots_adjust(wspace=0.5,hspace=0.5) # edit the gap manually, 0.5 = 50% of the axes width and height
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2) 
axes[0][1].plot(a,b)
axes[0][1].set_title("A vs B") # we can set title, xlabel for each subplots
fig.suptitle("Subplots", fontsize=16) # set a title for the entire Figure
axes[1][0].plot(x,y) 
plt.tight_layout() 
plt.show()
```

### Matplotlib Styling - Legends

```python
fig = plt.figure(figsize=(8,4), dpi=200) 
axes1 = fig.add_axes([0,0,1,1]) 
axes1.plot(a,b, label="A vs B") # create the label
axes1.plot(a,b*2, label="A vs B*2")
axes1.legend(loc="upper left") # show the legend and position with loc parameters
plt.show()
```

```python
fig = plt.figure(figsize=(8,4), dpi=200) 
axes1 = fig.add_axes([0,0,1,1]) 
axes1.plot(a,b, label="A vs B") # create the label
axes1.plot(a,b*2, label="A vs B*2")
axes1.legend(loc=(1.1, 0.5)) # show the legend and position with loc parameters outside the Figure
plt.show()
```

### Matplotlib Styling - Colors and Styles

```python
fig = plt.figure(figsize=(8,4), dpi=200) 
axes1 = fig.add_axes([0,0,1,1]) 
axes1.plot(a,b, label="A vs B", color="blue") # change the default color
axes1.plot(a,b*2, label="A vs B*2", color="#FF0099") # change the default color using hex color
axes1.legend(loc="best")
plt.show()
```

```python
fig = plt.figure(figsize=(8,4), dpi=200) 
axes1 = fig.add_axes([0,0,1,1]) 
axes1.plot(a,b, label="A vs B", color="blue", lw=3) # set line width 1 default value, 3 is 3 times the default
axes1.plot(a,b*2, label="A vs B*2", color="#FF0099", linewidth=0.5) 
axes1.legend(loc="best")
plt.show()
```

```python
fig = plt.figure(figsize=(8,4), dpi=200) 
axes1 = fig.add_axes([0,0,1,1]) 
axes1.plot(a,b, label="A vs B", color="blue", lw=3, ls="--") # set line style (dash)
axes1.plot(a,b*2, label="A vs B*2", color="#FF0099", linewidth=2, linestyle=":") # set line style (squared dot)
axes1.legend(loc="best")
plt.show()
```

```python
fig = plt.figure(figsize=(8,4), dpi=200) 
axes1 = fig.add_axes([0,0,1,1]) 
lines = axes1.plot(a,b, label="A vs B", color="blue", lw=3)
# set custom style. In each couple of numbers, the first refer to the number of dot, the second to the number of white space
lines[0].set_dashes([1,2,1,2,10,2]) 
axes1.legend(loc="best")
plt.show()
```

```python
fig = plt.figure(figsize=(8,4), dpi=200) 
axes1 = fig.add_axes([0,0,1,1]) 
lines = axes1.plot(a,b, label="A vs B", color="blue", lw=1, marker="+", markersize=10) # show the actual points we are plotting
axes1.legend(loc="best")
plt.show()
```

```python
fig = plt.figure(figsize=(8,4), dpi=200) 
axes1 = fig.add_axes([0,0,1,1]) 
lines = axes1.plot(a,b, label="A vs B", color="blue", lw=1, marker="o", markersize=10, 
                   markerfacecolor="red", markeredgewidth=4, markeredgecolor="orange") # cutomize the marker

axes1.legend(loc="best")
plt.show()
```

### Advanced Matplotlib Commands 

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
axes[0].plot(a,b) 
axes[0].set_title("Normal Scale")
axes[1].plot(a,b)
axes[1].set_title("Logarithmic Scale Y")
axes[1].set_yscale("log") # logarithmic scale for the y axes
plt.tight_layout() 
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(x, x**2, x, x**3, lw=2)

ax.set_xticks([0, 2, 4, 6, 8]) # we can position the ticks
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18) # set thiks labels

yticks = [0, 500, 1000, 1500]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.2f$" % y for y in yticks], fontsize=18); # use LaTeX formatted labels
```

```python
fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_title("scientific notation")

yticks = [0, 500, 1000, 1500]

# for larger number we can use the scientific notation
from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 
ax.yaxis.set_major_formatter(formatter) 
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,3))

# default grid appearance
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)

# custom grid appearance
axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color="blue")
ax1.set_ylabel(r"area $(m^2)$", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx() # with twinx or twiny we can create double axes plot
ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

```python
fig, ax = plt.subplots()

ax.spines['right'].set_color('none') # delete the right spine
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))   # set position of y spine to y=0

xx = np.linspace(-0.75, 1., 100)
ax.plot(xx, xx**3);
```

```python
fig, ax = plt.subplots()

ax.plot(xx, xx**2, xx, xx**3)

ax.text(0.15, 0.2, r"$y=x^2$", fontsize=20, color="blue") # we can add text annotation
ax.text(0.65, 0.1, r"$y=x^3$", fontsize=20, color="green");
```

```python
n = np.array([0,1,2,3,4,5])
```

```python
# we can create different type of plot
fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
```

### Distribution Plots

Distribution Plot display a single continuous variable and help visualize proprieties such deviation and average values

**Rug Plot**

A rug plot is the simplest distribution plot and merely adds a dash or tick line for evvery single value. The y-axis does not really have a meaning.

```python
sns.rugplot(data=df, x="salary", height=0.3)
plt.show()
```

**Histogram**

Histogram helps to visualize the frequency distribution of a continuos variable


```python
sns.displot(data=df, x="salary", bins=10)
plt.show()
```

```python
sns.set(style="darkgrid") # set the style for all the next notebook
sns.displot(data=df, x="salary", bins=10, color="orange", edgecolor="blue", linewidth=2)
plt.show()
```

```python
sns.displot(data=df, x="salary", bins=10, kde=True)# we can show the Kernel density estimation

plt.show()
```

```python
sns.displot(data=df, x="salary", bins=10, kde=True, rug=True)# we can also add the rug

plt.show()
```

```python
sns.histplot(data=df, x="salary", bins=10)# we can create a distribution plot also using hisplot
plt.show()
```

**KDE PLOT** 

KDE plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. KDE represents the data using a continuous probability density curve.


```python
sns.kdeplot(data=df, x="salary")
plt.show()
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=["age"])
```

```python
sns.displot(data=sample_ages, x="age", rug=True, bins=30, kde=True)
plt.show()
```

```python
sns.kdeplot(data=sample_ages, x="age", fill=True)
plt.show()
```

```python
sns.kdeplot(data=sample_ages, x="age", clip=[0,100], bw_adjust=0.1) 
plt.show()
```

The categorical plot shows statistical metrics per category. For exemple mean value for category

### Categorical Plots - Distributions within Categories

**Boxplot**

Boxplot shows the distribution of a continuous variabile using the quartiles. Quartiles separate out the data into equal number of data points: 25% of data points are in bottom quartile, 50% in the median, 75% in the top quartile.

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df, y="math score")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df, y="math score", x="test preparation course")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df, y="math score", x="parental level of education", hue="gender")
plt.legend(bbox_to_anchor=(1.05, 0.5))
plt.show()
```

**Violinplot**

The violinplot shows the propability density accross the data using a KDE. We can imagine it as a mirrored KDE plot.

```python
plt.figure(figsize=(10,5))
sns.violinplot(data=df, x="math score")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(data=df, x="math score", y="gender")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(data=df, x="math score", y="gender", hue="test preparation course", bw=0.1)
plt.legend(bbox_to_anchor=(1.3, 0.5))
plt.show()
```

**Swarmplot**

Swarmplot shows all the data points in the distribution

```python
plt.figure(figsize=(10,5))
sns.swarmplot(data=df, x="math score", y="gender")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(data=df, x="math score", y="gender", hue="test preparation course")
plt.legend(bbox_to_anchor=(1.3, 0.5))
plt.show()
```

**Boxenplot**

### Seaborn - Matrix Plots

Matrix plots are the visual equivalent of displaying a pivot table

**Heatmap**

Visually display the distribution of cell value with color mapping

```python
df = pd.read_csv("country_table.csv", index_col="Countries")
df.head()
```

```python
plt.figure(dpi=200)
sns.heatmap(df.drop("Life expectancy", axis=1), linewidths=0.5, annot=True, cmap="viridis", center=20)
plt.show()
```

**Clustermap**

Same visual as hitmap, but first conducts hierarchical clustering to reorganize data into groups

```python
plt.figure(dpi=200)
sns.clustermap(df.drop("Life expectancy", axis=1), linewidths=0.5, annot=True, cmap="viridis", center=20, col_cluster=False)
plt.show()
```
